```python
name = input("Enter student name: ")
marks = int(input("Enter student marks: "))

student = {
    "name": name,
    "marks": marks
}

print("Student Dictionary:", student)
```

```python
fruits = {
    "Apple": 120,
    "Banana": 40,
    "Mango": 100,
    "Orange": 80,
    "Grapes": 90
}

print("Fruit Dictionary:")
print(fruits)
```

```python
employees = {
    "Ravi": 50000,
    "Priya": 60000,
    "Amit": 55000
}

name = input("Enter employee name: ")

if name in employees:
    print("Salary:", employees[name])
else:
    print("Employee not found")
```

```python
countries = {
    "India": "New Delhi",
    "USA": "Washington D.C.",
    "Japan": "Tokyo",
    "France": "Paris"
}

print("Countries and Capitals:")

for country, capital in countries.items():
    print(country, ":", capital)
```

```python
student = {
    "name": "Rahul",
    "marks": 75
}

new_marks = int(input("Enter new marks: "))

student["marks"] = new_marks

print("Updated Dictionary:")
print(student)
```

```python
inventory = {
    "Pen": 50,
    "Book": 30,
    "Pencil": 100
}

product = input("Enter product name: ")
amount = int(input("Enter quantity to add: "))

if product in inventory:
    inventory[product] += amount
    print("Updated Inventory:", inventory)
else:
    print("Product not found")
```

```python
student = {
    "name": "Rahul",
    "marks": 75,
    "age": 20
}

key = input("Enter key to delete: ")

if key in student:
    del student[key]
    print("Updated Dictionary:", student)
else:
    print("Key not found")
```

```python
student = {
    "name": "Rahul",
    "marks": 75,
    "age": 20
}

student.clear()

print("Dictionary after removing all items:")
print(student)
```
