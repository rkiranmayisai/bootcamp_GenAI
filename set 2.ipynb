```python
sentence = "python is easy and python is powerful"

words = sentence.split()
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print("Word Frequencies:")
print(frequency)
```

```python
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 15, 'c': 25, 'd': 40}

merged = dict1.copy()

for key, value in dict2.items():
    if key in merged:
        merged[key] += value
    else:
        merged[key] = value

print("Merged Dictionary:")
print(merged)
```

```python
student_marks = {
    'Alice': 85,
    'Bob': 92,
    'Charlie': 78,
    'David': 95
}

sorted_dict = sorted(student_marks.items(), key=lambda x: x[1], reverse=True)

print("Dictionary sorted by values (descending):")
for key, value in sorted_dict:
    print(key, ":", value)
```

```python
scores = {
    'A': 90,
    'B': 95,
    'C': 95,
    'D': 80
}

max_value = max(scores.values())

max_keys = []

for key, value in scores.items():
    if value == max_value:
        max_keys.append(key)

print("Maximum Value:", max_value)
print("Key(s) with Maximum Value:", max_keys)
```

```python
original = {
    'A': 1,
    'B': 2,
    'C': 1,
    'D': 3,
    'E': 2
}

inverted = {}

for key, value in original.items():
    if value not in inverted:
        inverted[value] = [key]
    else:
        inverted[value].append(key)

print("Inverted Dictionary:")
print(inverted)
```
